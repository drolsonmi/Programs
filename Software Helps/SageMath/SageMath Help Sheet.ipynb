```python
# Basic plotting functions

g = x^2 + 3
y = 3*(x+1)^2+2
plot(g,(x,0,1), ymin=-1,ymax=10,legend_label="$x^2 + 3$", title="A Function of $y(x)$", axes_labels=['$x$', '$y$']) + plot(y, (-1,2), color='red', legend_label="$3(x+1)^2+2$")
```

```python
# Plotting as objects

f = piecewise([((0,1), x^3), ([-1,0], -x)]);
g = -3*x^2 + 3

p = plot(f,legend_label="piecewise",linestyle="-.")
#p.set_axes_range(xmin=-10, xmax=2.5, ymin=-1.0, ymax=4.5)

q = plot(g,(x,-0.5,0.3), color='red', legend_label="$-3x^2+3$", legend_color="green", linestyle=":")

p + q
# or p.plot() + q.plot()

#p.save("plot2.svg", xmin=-3, xmax=3)
```

```python
f = piecewise([((0,1), x^3), ([-1,0], -x)]);
g = -3*x^2 + 3

p = plot(f,legend_label="piecewise") + plot(g,(x,-0.5,0.3), color='red', legend_label="$-3x^2+3$", legend_color="green")
p.show(xmin=-2.5,xmax=0.75,ymin=-1.0,ymax=4.5)
```

```python
# Adding points, lines, and other features to the graph

a = 1.0
f(x) = x^2 * sin(1/x)

g1 = plot(f(x), (x, -2, 2), , title="Test figure")
g1 += line([(a, 0), (a, f(a))], color='green')
g1 += point(((0, 0), (a, f(a))), color='red', pointsize=20)
g1 += text('A', (-0.05, 0.1), color='red')
g1 += text('B', (a + 0.05, f(a) - 0.1), color='green')
g1 += text ('Alternate Title', (0.5,2.0), color='black')
g1 += polygon([(1.25,0),(1.25,f(2))] + [(x, f(x)) for x in [1.25,1.3,..,2]] + [(2,0),(1.25,0)],  rgbcolor=(0.8,0.8,0.8),aspect_ratio='automatic')

g2 = plot(f(x), (x, -0.3, 0.3), axes_labels=['$x$', '$y$'], frame=True, axes=True)

g1.inset(g2, pos=(0.15, 0.7, 0.25, 0.25))
```

## Vectors

```python
v = vector([1,2,3])      # Vector of Integers
v = vector(QQ,[1,2,3])   # Vector of Real Numbers

# Dot Product
u = vector(QQ,[1/2,1/3,1/4])
u.dot_product(v)

# Cross Product
u.cross_product(v)

# Inner Product
u.inner_product(v)
```
