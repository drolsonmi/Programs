```python
from math import *
from pylab import *
#%pylab inline
%matplotlib inline

def function(x):
    y = exp(x)
    return y

xmin = 0
xmax = 6
ymin = 0
ymax = 100
```

```python
x = arange(xmin,xmax,0.005)
y = function(x)
plot(x,y,'b')
ylim([ymin,ymax])
title("Hello World")
```
